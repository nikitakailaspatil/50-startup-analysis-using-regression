```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("50_Startups.csv")
```

```python
df.head() #it will print first 5 records
```

```python
#total 5 columns
#toatal entries=50
#No null values
```

```python
#column seperation of numeric columns
num_df=df.select_dtypes(["float64"])
```

```python
num_df
```

```python
#seperation of categorical columns
cat_df=df.select_dtypes(["object"])
```

```python
cat_df
```

```python
#label encoding
#import module
from sklearn.preprocessing import LabelEncoder
#create an object of LabelEncoder
le=LabelEncoder()
#Using fit_transform to perform label encoding
for i in cat_df:
    le=LabelEncoder()
    cat_df[i]=le.fit_transform(cat_df[i])
```

```python
cat_df
```

```python
df2=pd.concat([num_df,cat_df],axis=1)
```

```python
df2
```

```python
#Seperate data into x and y
x=df2[["R&D Spend","Administration","Marketing Spend","State"]]
x
```

```python
y=df2[["Profit"]]
y
```

```python
#split data
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
#model creation
from sklearn.linear_model import LinearRegression
le=LinearRegression()
le.fit(xtrain,ytrain)
ypred=le.predict(xtest)
```

```python
ypred
```

```python
#performance metric
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(f"MAE={mean_absolute_error(ytest,ypred)}")
print(f"MSE={mean_squared_error(ytest,ypred)}")
print(f"RMSE={np.sqrt(mean_squared_error(ytest,ypred))}")
print(f"ACCURACY={r2_score(ytest,ypred)}")
```
